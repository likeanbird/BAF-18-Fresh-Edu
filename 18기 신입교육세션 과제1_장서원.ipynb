## 1번 과제 관심 있는 주제 데이터 찾기
관심있는 주제: 우리나라 폐암 발생 현황과 주요 원인 \
이유: 최근 국민건강보험공단이 담배 제조사를 상대로 손해배상 소송을 청구했다는 기사를 접했는데, 실제로 흡연이 폐암을 유발하는 데 얼마나 결정적인 영향을 끼치는지 궁금했기 때문이다.

- 데이터명: 국립암센터_암발생 통계 정보
- 데이터 특징
   - 출처: 공공데이터포털(https://www.data.go.kr/data/3039563/fileData.do)
   - 분야: 보건 의료
   - 내용: 중앙암등록본부가 국가암등록통계사업을 통해 수집한 1999년부터 2022년까지의 대한민국 암 발생 현황 자료이다.
   - 성별(남/여/전체), 연령대(0세부터 5세 단위로 분류), 암종별에 따른 연도별 암 발생 통계를 제공하고 있다.
   - 핵심 지표는 발생자 수와 조발생률(10만 명당 발생률)로 보여진다. 각 분류체계에 따른 암 발생 추세 분석 등에 활용할 수 있을 것으로 보인다.


- 데이터명: Lung Cancer
- 데이터 특징
   - 출처: kaggle(https://www.kaggle.com/datasets/mysarahmadbhat/lung-cancer/data)
   - 분야: 보건 의료
   - 내용: 온라인 폐암 예측 시스템으로 수집된 284명의 자료이다.
   - 폐암과 관련된 생활 습관(smoking/alcohol), 신체 반응(yellow fingers/wheezing/coughing 등)을 비롯한 16가지 지표를 포함하고 있는 자료이다.
   - 수집된 데이터에 국적 정보가 없다는 점을 고려하여, 인종이나 환경에 따른 차이가 결과에 미칠 수 있는 영향 역시 염두에 둬야 할 것으로 보인다.
