### 성별과 생존 여부는 독립이다 vs 아니다
범주형과 범주형 자료의 관계 -> 카이제곱 검정

```python
from scipy.stats import chi2_contingency # 카이제곱 검정 함수 불러오기

table_sex = pd.crosstab(data1['Sex'], data1['Survived']) # 교차표 생성하기(결과는 각 조합의 빈도 표)
chi2, p, dof, expected = chi2_contingency(table_sex)
# chi2: 카이제곱 통계량, p: 유의확률, dof: 자유도, expected: 귀무가설 하에 기대되는 빈도

print(chi2, p) # 통계량, p-value 출력하기
```

```python
p<0.05 # 유의확률 0.05이하 여부 확인
```

성별과 생존 여부는 독립이라는 귀무가설을 기각한다. -> 성별과 생존 여부는 관련이 있다.

### 객실등급과 생존 여부는 독립이다 vs 아니다
범주형과 범주형 자료의 관계 -> 카이제곱 검정

```python
table_pclass = pd.crosstab(data1['Pclass'], data1['Survived'])
chi2, p, dof, expected = chi2_contingency(table_pclass)

print(chi2, p)
```

```python
p<0.05
```

객실 등급과 생존 여부는 독립이라는 귀무가설을 기각한다. -> 객실 등급은 생존 가능성과 관련이 있다.

### 가족규모와 생존 여부는 독립이다 vs 아니다
범주형과 범주형 자료의 관계 -> 카이제곱 검정

```python
table_family = pd.crosstab(data1['family'], data1['Survived'])
chi2, p, dof, expected = chi2_contingency(table_family)

print(chi2, p)
```

```python
p<0.05
```

가족규모과 생존 여부는 독립이라는 귀무가설을 기각한다. -> 가족규모 생존 가능성과 관련이 있다.

### 생존자와 사망자의 평균 나이가 같다 vs 아니다
연속형(age)와 이진 범주(생존여부) -> welch t-test 활용
아직 결측치를 처리하기 전이어서 na값은 제거하고 테스트 진행

```python
from scipy.stats import ttest_ind

age_survived = data1.loc[data1['Survived'] == 1, 'Age'].dropna()
age_dead = data1.loc[data1['Survived'] == 0, 'Age'].dropna()

ttest_ind(age_survived, age_dead, equal_var=False)
```

```python
0.04118965162586639<0.05
```

생존자와 사망자의 평균 나이에 대해 t-test를 수행한 결과, 두 집단 간 평균 차이는 통계적으로 유의함을 확인할 수 있었다. 이는 나이 변수가 생존 여부과 관련이 있음을 확인할 수 있다.

```python
data1['Age_title'] = data1['Age'] # title을 활용한 age 결측치 채우기
data1.head(1)
```

```python
age_na = data1['Age'].isnull() # 결측치가 존재하는 행 뽑아내기
age_na
```

```python
data1.loc[age_na, 'Age_title'] = data1.loc[age_na, 'Title'].map(title_age_median)
# title 중앙값으로 결측치 채우기
# data1.loc[age_na, 'Title'] -> age가 결측이 사람의 title 추출
# .map(title_age_median) -> 각 title 중앙값으로 변환하
```

```python
data1['Age_title'].isnull().sum()
```

```python
### 생존자와 사망자의 평균 나이가 같다 vs 아니다
```

```python
from scipy.stats import ttest_ind

# 생존자와 사망자의 Age_title 분리
age_title_survived = data1.loc[data1['Survived']==1, 'Age_title'] # 생존자 집단의 나이
age_title_dead = data1.loc[data1['Survived']==0, 'Age_title'] # 사망자 집단의 나이

# Welch t-test 수행(분산이 알려져 있지 않아 -> 같지 않다고 가정함)
ttest_ind(age_title_survived, age_title_dead, equal_var=False)
```

title을 기준으로 결측치를 대체한 age_title을 이용해 t-test를 진행한 결과, 생존자와 사망자의 평균 차이는 p-value<0.05으로 통계적으로 유의한 것을 확인할 수 있었다.
결측치 처리를 하기 전과 비교했을 때 표본 수와 분산이 변화하며 검정의 민감도가 달라진 것으로 보인다. 다만, 귀무가설을 기각한다느 결론 자체는 유지되었다.

## 2. Sex와 Pclass를 기준으로 결측치 채워보기

```python
age_mean = data1.groupby(['Sex', 'Pclass'])['Age'].mean() # sex와 pclss 기준 평균 나이 계산

data1['Age_sp'] = data1['Age'] # 복사본

age_na2 = data1['Age'].isnull() 

data1.loc[age_na2, 'Age_sp'] = data1.loc[age_na2, ['Sex','Pclass']].apply(
    lambda x: age_mean.loc[(x['Sex'], x['Pclass'])], axis=1) # sex와 pclass 그룹의 평균 나이로 대체하기

data1['Age_sp']
```

```python
from scipy.stats import ttest_ind

age_title_survived = data1.loc[data1['Survived']==1, 'Age_sp'] # 생존자 집단의 나이
age_title_dead = data1.loc[data1['Survived']==0, 'Age_sp'] # 사망자 집단의 나이

ttest_ind(age_title_survived, age_title_dead, equal_var=False)
```

age_title과 동일하게 age_sp를 사용하여 t-test를 수행한 결과, 생존자와 사망자의 평균 차이는 p-value<0.05으로 통계적으로 유의한 것을 확인할 수 있었다.
