```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=train['Age'])
plt.title('ë‚˜ì´ ë³€ìˆ˜ ì´ìƒì¹˜ íƒìƒ‰ (Boxplot)')
plt.xlabel('ë‚˜ì´')
plt.show()

# IQR(ì‚¬ë¶„ìœ„ê°„ ë²”ìœ„) ê³„ì‚°
Q1 = train['Age'].quantile(0.25)
Q3 = train['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_age = Q1 - 1.5 * IQR
upper_age = Q3 + 1.5 * IQR

# ê²°ê³¼ ì¶œë ¥
print(f"< ë‚˜ì´(Age) í†µê³„ ìš”ì•½ >")
print(f"í•˜í•œì„ : {lower_age}")
print(f"ìƒí•œì„ : {upper_age}")
print(f"ìµœì†Ÿê°’: {train['Age'].min()}")
print(f"ìµœëŒ“ê°’: {train['Age'].max()}")

outliers_age = train[(train['Age'] < lower_age) | (train['Age'] > upper_age)]
print(f"ì´ìƒì¹˜ ë°ì´í„° ê°œìˆ˜: {len(outliers_age)}ê°œ")
```

```python
import numpy as np
from scipy.stats import chi2_contingency

def cramers_v(confusion_matrix):
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    # í¸í–¥ ë³´ì • ê³µì‹ ì ìš©
    phi2_corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    r_corr = r - ((r-1)**2)/(n-1)
    k_corr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2_corr / min((k_corr-1), (r_corr-1)))

# ì´ì „ì˜ êµì°¨í‘œ(contingency_table) ì‚¬ìš©
v_score = cramers_v(contingency_table)
print(f"Cramer's V: {v_score:.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, classification_report, 
                             mean_squared_error, r2_score)

# ë°ì´í„° ì¤€ë¹„ (ì´ì „ ì„¤ì • ìœ ì§€)
exclude_cols = ['CustomerID', 'Subscription Type', 'AgeGroup', 'Tenure_Group', 
                'Usage_Density', 'Spend_Efficiency', 'Active_Life_Span', 'Delay_per_Usage', 'Subscription Type_Basic','Subscription Type_Standard','Subscription Type_Premium', 'Contract Length_Monthly','Churn']
X = train.drop(columns=[col for col in exclude_cols if col in train.columns])
y = train['Churn']

# ë²”ì£¼í˜• ë³€ìˆ˜ ì¸ì½”ë”© (Gender, Contract Length ë“±)
X_encoded = pd.get_dummies(X, drop_first=True)

# ë°ì´í„° ë¶„í•  (í•™ìŠµìš© 80%, í…ŒìŠ¤íŠ¸ìš© 20%)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)
```

### Logistic Regression

```python
# ì „ì²˜ë¦¬: í‘œì¤€í™”í•˜ê¸°
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ëª¨ë¸ í•™ìŠµ
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_scaled, y_train)

# ì˜ˆì¸¡ ë° í™•ë¥  ì‚°ì¶œ
y_pred = lr_model.predict(X_test_scaled)
y_prob = lr_model.predict_proba(X_test_scaled)[:, 1] # ì´íƒˆ í™•ë¥ ê°’

# RMSE, R2
mse = mean_squared_error(y_test, y_prob)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_prob)

print("--- Logistic Regression í‰ê°€ ê²°ê³¼ ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R-Square: {r2:.4f}")
print("\n[ìƒì„¸ ë¦¬í¬íŠ¸]")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

# [1] ë°ì´í„° ë¡œë“œ
train = pd.read_csv('customer_churn_train.csv')
test = pd.read_csv('customer_churn_test.csv')

# [2] í†µí•© ì „ì²˜ë¦¬ í•¨ìˆ˜ (ì‚¬ìš©ìì˜ íŒŒìƒ ë³€ìˆ˜ ë¡œì§ ë°˜ì˜)
def preprocess_data(df):
    df = df.copy()
    
    # 1. ê²°ì¸¡ì¹˜ ì„ ì œ ì²˜ë¦¬
    df = df.fillna(0)
    
    # 2. ì‚¬ìš©ìê°€ ì •ì˜í•œ íŒŒìƒ ë³€ìˆ˜ ìƒì„±
    df['Senior'] = (df['Age'] >= 50).astype(int)
    df['Support_per_Usage'] = df['Support Calls'] / (df['Usage Frequency'] + 1)
    df['Value_per_Support'] = df['Total Spend'] / (df['Support Calls'] + 1)
    df['Usage_Density'] = df['Usage Frequency'] / (df['Tenure'] + 1)
    df['Inactivity_Ratio'] = df['Last Interaction'] / (df['Tenure'] + 1)
    df['Active_Life_Span'] = df['Tenure'] - df['Last Interaction']
    df['Delay_per_Usage'] = df['Payment Delay'] / (df['Usage Frequency'] + 1)
    df['Spend_Efficiency'] = df['Total Spend'] / (df['Tenure'] + 1)
    df['IsMonthly'] = (df['Contract Length'] == 'Monthly').astype(int)
    
    # 3. ëª¨ë¸ë§ì—ì„œ ì œì™¸í•  ë³€ìˆ˜ ë¦¬ìŠ¤íŠ¸ (ì‚¬ìš©ì ì§€ì • ê¸°ì¤€)
    exclude_cols = [
        'CustomerID', 'Subscription Type', 'AgeGroup', 'Tenure_Group', 
        'Usage_Density', 'Spend_Efficiency', 'Active_Life_Span', 'Delay_per_Usage'
    ]
    
    # 4. ë³€ìˆ˜ ì œê±° ë° ì¸ì½”ë”©
    target = 'Churn' if 'Churn' in df.columns else None
    X = df.drop(columns=[col for col in exclude_cols if col in df.columns])
    if target:
        X = X.drop(columns=[target])
        
    X_enc = pd.get_dummies(X, drop_first=True)
    
    # 5. ë¬´í•œëŒ€(inf) ë° ê²°ì¸¡ì¹˜ ìµœì¢… ë°©ì–´ (ValueError í•´ê²°)
    X_enc = X_enc.replace([np.inf, -np.inf], np.nan).fillna(0)
    
    return X_enc, (df['Churn'] if target else None)

# [3] ì „ì²˜ë¦¬ ì‹¤í–‰
X_train, y_train = preprocess_data(train)
X_test, y_test = preprocess_data(test)

# [4] ì»¬ëŸ¼ ìˆœì„œ ë° êµ¬ì¡° ì¼ì¹˜í™” (ë§ˆì´ë„ˆìŠ¤ R2 í•´ê²°)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# [5] LightGBM ëª¨ë¸ í•™ìŠµ (NotFittedError í•´ê²°)
lgbm_model = LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=42, verbose=-1)
lgbm_model.fit(X_train, y_train)

# [6] ì„±ëŠ¥ í‰ê°€ ë° ì˜ˆì¸¡
def evaluate_model(model, X, y, label):
    if y is not None:
        probs = model.predict_proba(X)[:, 1]
        rmse = np.sqrt(mean_squared_error(y, probs))
        r2 = r2_score(y, probs)
        acc = accuracy_score(y, model.predict(X))
        print(f"[{label}] Accuracy: {acc:.4f} | RMSE: {rmse:.4f} | R-Square: {r2:.4f}")

evaluate_model(lgbm_model, X_train, y_train, "Train Set")
evaluate_model(lgbm_model, X_test, y_test, "Test Set")

# [7] ìµœì¢… ê²°ê³¼ ì €ì¥
test_preds = lgbm_model.predict(X_test)
output = pd.DataFrame({'CustomerID': test['CustomerID'], 'Churn': test_preds})
output.to_csv('test.csv', index=False)
print("\nìµœì¢… ì˜ˆì¸¡ ê²°ê³¼ê°€ 'test.csv'ë¡œ ì €ì¥ë˜ì—ˆë‹¤.")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

# [1] ë°ì´í„° ë¡œë“œ
train = pd.read_csv('customer_churn_train.csv')
test = pd.read_csv('customer_churn_test.csv')

# [2] í†µí•© ì „ì²˜ë¦¬ í•¨ìˆ˜ (ê²°ì¸¡ì¹˜ ë° ë¬´í•œëŒ€ ì™„ì „ ë°©ì–´)
def rf_preprocess(df):
    df = df.copy()
    
    # 1. ê¸°ì´ˆ ê²°ì¸¡ì¹˜ ì²˜ë¦¬ (NaN ì „ì´ ë°©ì§€)
    df = df.fillna(0)
    
    # 2. Gender ì¸ì½”ë”© (Female: 2, Male: 1)
    if df['Gender'].dtype == 'object':
        df['Gender'] = df['Gender'].map({'Female': 2, 'Male': 1}).fillna(1)
    
    # 3. íŒŒìƒ ë³€ìˆ˜ ìƒì„± (ê¸°ì¡´ ë¡œì§ ìœ ì§€)
    df['Senior'] = (df['Age'] >= 50).astype(int)
    df['Support_per_Usage'] = df['Support Calls'] / (df['Usage Frequency'] + 1)
    df['Value_per_Support'] = df['Total Spend'] / (df['Support Calls'] + 1)
    df['Inactivity_Ratio'] = df['Last Interaction'] / (df['Tenure'] + 1)
    df['IsMonthly'] = (df['Contract Length'] == 'Monthly').astype(int)
    
    # 4. ë¶„ì„ ì œì™¸ ë³€ìˆ˜ ì„¤ì •
    exclude_cols = [
        'CustomerID', 'Subscription Type', 'AgeGroup', 'Tenure_Group', 
        'Usage_Density', 'Spend_Efficiency', 'Active_Life_Span', 'Delay_per_Usage'
    ]
    
    # 5. ë³€ìˆ˜ ì„ íƒ ë° ë”ë¯¸í™”
    target = 'Churn' if 'Churn' in df.columns else None
    X = df.drop(columns=[col for col in exclude_cols if col in df.columns])
    if target: X = X.drop(columns=[target])
    
    X_enc = pd.get_dummies(X, drop_first=True)
    
    # 6. ë…¸íŠ¸ë¶ ìˆ˜ë™ ë“œë ì»¬ëŸ¼ ì²˜ë¦¬
    extra_drop = ['Subscription Type_Basic', 'Subscription Type_Standard', 
                  'Subscription Type_Premium', 'Contract Length_Monthly']
    X_enc = X_enc.drop(columns=[col for col in extra_drop if col in X_enc.columns])
    
    # 7. â˜…ìµœì¢… ë¬´í•œëŒ€(inf) ë° ê²°ì¸¡ì¹˜ ì œê±°â˜… (ValueError ë°©ì–´ í•µì‹¬)
    X_enc = X_enc.replace([np.inf, -np.inf], np.nan).fillna(0)
    
    return X_enc, (df['Churn'] if target else None)

# [3] ì „ì²˜ë¦¬ ì‹¤í–‰
X_train, y_train = rf_preprocess(train)
X_test, y_test = rf_preprocess(test)

# [4] ì»¬ëŸ¼ ìˆœì„œ ì¼ì¹˜í™” (í•™ìŠµ ë°ì´í„° ê¸°ì¤€)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# [5] ëœë¤ í¬ë ˆìŠ¤íŠ¸ ëª¨ë¸ í•™ìŠµ
# ê³¼ì í•© ë°©ì§€ë¥¼ ìœ„í•´ max_depthë¥¼ ì„¤ì •í•˜ê³ , n_jobs=-1ë¡œ ë³‘ë ¬ ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•œë‹¤.
rf_model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)

# [6] ì„±ëŠ¥ í‰ê°€ í•¨ìˆ˜
def show_performance(model, X, y, label):
    if y is not None:
        probs = model.predict_proba(X)[:, 1]
        rmse = np.sqrt(mean_squared_error(y, probs))
        r2 = r2_score(y, probs)
        acc = accuracy_score(y, model.predict(X))
        print(f"[{label}] Accuracy: {acc:.4f} | RMSE: {rmse:.4f} | R-Square: {r2:.4f}")

show_performance(rf_model, X_train, y_train, "Train Set")
show_performance(rf_model, X_test, y_test, "Test Set (Final)")

# [7] ê²°ê³¼ ì €ì¥
test_preds = rf_model.predict(X_test)
output = pd.DataFrame({'CustomerID': test['CustomerID'], 'Churn': test_preds})
output.to_csv('test.csv', index=False)
```
